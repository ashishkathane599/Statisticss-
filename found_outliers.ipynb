```python
import matplotlib.pyplot as plt 
import numpy as np 
%matplotlib inline 
```

```python
# define a dataset 
dataset = [11,10,12,14,12,15,14,13,15,102,12,14,17,19,10,13,12,14,13,15,10,15,12,10,14,13,15,102,107,108]
```

 ### Z-score  Method :- to find outliers 

```python
# Z-score 
outliers = [ ] 
def detect_outliers(data) : 
    threshold=3  # 3rd standard daviation
    std_min = 1 
    mean = np.mean(data) 
    std = np.std(data) 
    for i in   data : 
        z_score = (i-mean)/std
        if  np.abs(z_score) > threshold  :  # abs give a positive value 
            outliers.append(i) 
    return outliers 
```

```python
detect_outliers(dataset)
```

### 2. IQR :- technique to identify the outliers 
This method is more efficient then the z-score  method 
Steps: 
1. Sort the data
2. Calculate the Q1(25%) and Q3(75%)
3. IQR(Q3-Q1)
4. Find the lower Fence(Q1-1.5(IQR))
5. Find the upper Fence( Q3 + 1.5(IQR)) 

```python
#sort 
dataset = sorted(dataset) 
```

```python
q1,q3 = np.percentile(dataset,[25,75])   # calculating q1,q2 
q1,q3
```

```python
iqr = q3-q1 
iqr
```

```python
# find the lower fence and upper fence 
lower_fence = q1-(1.5*iqr)   # bellow elemets are outliers 
upper_fence = q3+(1.5*iqr)   # above elements are outliers 
```

```python
outliers =[ ]
for i in dataset : 
    if  lower_fence< i < upper_fence : 
        pass
    else : 
       outliers.append(i) 
print("outliers : " , outliers )
```
