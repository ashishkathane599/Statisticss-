# Measure Of Dispersion 
1. Variance
2. Standard Deviation 

```python
statistics.variance(ages)   # use sample variable formula 
```

```python
statistics.pvariance(ages)  # for population variance 
```

```python
np.var(ages,axis=0)    # it is a population variance 
```

```python
# manualy var calculation 
def variance(data) : 
    n = len(ages) 
    # mean of data 
    mean = sum(data)/n
    #variance 
    deviation = [(x-mean)**2 for x in data] 
    variance = sum(deviation) / n   # for sample var n-1 
    return variance   
    
```

```python
variance(ages)
```

```python
# for population and sample variance 
def variance(data,dof = 0 ) : 
    n = len(data) 
    # mean of data 
    mean = sum(data) / n 
    # variance 
    daviation = [(x-mean)**2 for x in data]
    variance =  sum(daviation) / (n-dof)   # if dof =0 then it is population variance
    return variance 
   # and if dof is 1 then it is sample variance 
```

```python
variance(ages,dof = 1 )
```

## Standard daviation 

```python
# for standard daviation 
import math
math.sqrt(statistics.pvariance(ages))  # population standard daviation 
```

```python
math.sqrt(statistics.variance(ages))  # sample standard daviation 
```

# Check whether Dis}tribution is Normal Distribution 
## * Q-Q  Plot 

```python
# If you want to check Whether feature is guassian or normal distribution
# Q-Q Plot 
import matplotlib.pyplot as plt 
import scipy.stats as stat 
import pylab 
def plot_data(sample) : 
    plt.figure(figsize=(10,6)) 
    plt.subplot(1,2,1) 
    sns.histplot(sample) 
    plt.subplot(1,2,2) 
    stat.probplot(sample,dist = 'norm',plot = pylab) 
    plt.show() 
```

```python
# create a normal distributed data 
s = np.random.normal(0.5,0.2,1000) 
plot_data(s) 

# IF LINE IT GIVES A STRAIGHT LINE THEN IT IS NORMAL DISTRIBUTIION 
```

```python
plot_data(df['sepal_width'])
```

```python
# for log normal distribution 
# Q-Q Plot 
import matplotlib.pyplot as plt 
import scipy.stats as stat 
import pylab 
def plot_data_log(sample) : 
    plt.figure(figsize=(10,6)) 
    plt.subplot(1,2,1) 
    sns.histplot(sample) 
    plt.subplot(1,2,2) 
    stat.probplot(np.log(sample),dist = 'norm',plot = pylab) 
    plt.show() 
```
